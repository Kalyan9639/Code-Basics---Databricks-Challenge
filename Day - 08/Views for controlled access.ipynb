# Views in Databricks:

A view in Databricks is:

- A virtual table defined by a SQL query over one or more tables or other views.
- **Read-only**: You cannot directly change data in a view; any changes must be made to the underlying tables.
- Used for **controlled access**: You can restrict which rows or columns users see, or mask sensitive data, by defining the viewâ€™s query appropriately (for example, using dynamic views for row/column-level access control).
- **Does not store data**: It only stores the query definition, not a copy of the data. Querying a view runs the underlying query each time.

```python
%sql
insert into example.bronze.custom_table values 
(3, 'John', 25),
(4, 'Jane', 30),
(5, 'Bob', 35);
```

```python
%sql
create view example.bronze.custom_view as select name,age from example.bronze.custom_table;
```
