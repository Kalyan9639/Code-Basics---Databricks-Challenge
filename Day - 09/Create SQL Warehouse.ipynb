### Craeting a new table and copying the contents of `custom_table` to it

```python
%sql
create table if not exists example.bronze.table_with_sql
(
  id int primary key not null,
  name string,
  age int,
  salary int
);
-- use example.bronze.table_with_sql;

insert into example.bronze.table_with_sql 
select id,name,age,NULL as salary from example.bronze.custom_table;
```

```python
%sql
select * from example.bronze.table_with_sql;
```

```python
%sql
-- fill the salary column with appropriate values
update example.bronze.table_with_sql
set salary= CASE
  when id = 1 then 999999999
  when id = 2 then 999999998
  when id = 3 then 22598746
  when id = 4 then 3786765
  when id = 5 then 2346635
  else salary
end
where id in (1,2,3,4,5);

select * from example.bronze.table_with_sql
```
